```python
from PIL import Image
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame()
```

```python
def pixel(pix_val):
    pixlist=[]
    for i in range(0,len(pix_val)):
        if(pix_val[i][0]==pix_val[i][1]==pix_val[i][2]==255):
            pixlist.append(0)
        else:
            pixlist.append(1)
    return pixlist
```

```python
import os, os.path

for root, dirs, files in os.walk('.'):
    print(root)
    if(root=='.\test-images\0'):
        print("yes")
    for f in files:
        fullpath = os.path.join(root, f)
        if os.path.splitext(fullpath)[1] == '.png':
            print(fullpath)
            image2=Image.open(fullpath)
            image2.show()
            pix_val = list(image2.getdata())
            pilist=pixel(pix_val)
            df1 = pd.DataFrame(np.array(pilist).reshape(-1,len(pilist)))
            print(df1)
            df=df.append(df1,ignore_index=True)
            df1=pd.DataFrame()
            
            
            #image2.show()
    
        
```

```python
df['Class']='Zero'
```
