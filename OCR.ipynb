#                                              OCR

# Importing Data

```python
import pandas as pd
import numpy as np
dt =pd.read_csv('OCRtrain.csv')
```

```python
dt
```

```python
X=dt.iloc[:,0:784]
y=dt.ix[:,784]
```

# Image-Processing

```python
from PIL import Image
import pandas as pd
import numpy as np
```

```python
image1=Image.open('thr.png')
image1.show()
```

```python
pix_val = list(image1.getdata())
```

```python
def pixel(pix_val):
    pixlist=[]
    for i in range(0,len(pix_val)):
        if(pix_val[i][0]==pix_val[i][1]==pix_val[i][2]==255):
            pixlist.append(0)
        else:
            pixlist.append(1)
    return pixlist
```

```python
pixlist=pixel(pix_val)
print(pixlist)
```

```python
pix=np.asarray(pixlist)
```

# Random-Forest

```python
parameters={'n_estimators':[10,20,50,100],'criterion':['gini','entropy'],'max_depth':[4,5,6,7,8,9],'min_samples_split':[5,10,15]}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
det = RandomForestClassifier(random_state=42)
clf = GridSearchCV(det, parameters, cv=5)
clf.fit(X,y)
```

```python
print(clf.best_estimator_)
```

```python
print(clf.best_score_ )
```

Predicted Image using Random Forest Algorithm

```python
val=clf.predict([pix])
print(val)
```

# SVM

```python
parameters={'C':[1,10,100],'kernel':['rbf'],'degree':[2,3]}
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC(random_state=42,gamma=0.01)
clf = GridSearchCV(svm, parameters, cv=5)
clf.fit(X,y)

```

```python
print(clf.best_estimator_)
```

```python
print(clf.best_score_)
```

Predicted Image using SVM Algorithm

```python
val=clf.predict([pix])
print(val)
```
